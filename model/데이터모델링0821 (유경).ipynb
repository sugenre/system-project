### XG Boost 모델

```python
import xgboost as xgb
from xgboost.sklearn import XGBRegressor

#3가지 초모수 튜닝 (max_depth, learning_rate, subsample)

hr2_accuracy_list = []

max_depth_list = [3,5,6] #기본값 6
learning_rate_list = [0.03, 0.07, 0.3] #Eta 0.3
subsample_list = [0.7, 0.9, 1.0] #기본값 1

def hr2_grid_search(max_depth,learning_rate, subsample): 
    xgb_model = xgb.XGBRegressor(max_depth=max_depth,learning_rate = learning_rate, subsample=subsample,seed=11)
    xgb_model.fit(X_train, Y_train)
    y_pred = xgb_model.predict(X_test)
    
    return([max_depth, learning_rate, subsample, xgb_model.score(X_train, Y_train), xgb_model.score(X_test, Y_test)])
      
for max_depth in max_depth_list:
    for learning_rate in learning_rate_list:
        for subsample in subsample_list:
            hr2_accuracy_list.append(hr2_grid_search(max_depth, learning_rate, subsample))
    
hr2_xgb_df = pd.DataFrame(hr2_accuracy_list, columns=['max_depth','learning_rate', 'subsample rate', 'train score','test score'])
hr2_xgb_df.sort_values('test score', ascending=False) # test값 내림차순으로 출력
```

```python
xg_model = xgb.XGBRegressor(max_depth=5, learning_rate=0.3, subsample=1.0, seed=11)
xg_model.fit(X_train, Y_train, verbose =False)
y_pred = xg_model.predict(X_test)

print("훈련 세트 정확도 (r2): {:.3f}".format(xg_model.score(X_train, Y_train)))
print("테스트 세트 정확도 (r2): {:.3f}".format(xg_model.score(X_test, Y_test)))
```
